```python
df = pd.read_csv("../data/preprocessed/colex_pron_geo_dedup.csv")
```

```python
def diff(colex):
    t1, t2 = colex.split("~")
    if t1!=t2:
        return True
    else: return False
```

```python
df["diff"]=df["COLEX"].apply(lambda x: diff(x))
```

```python
langs = list(set(df["LANG_PRON"]))
colexes = list(set(df["COLEX"]))
```

```python
len(langs), len(colexes)
```

```python
from collections import defaultdict
```

```python
lang_colex= defaultdict(dict)

for lang in tqdm(langs):
    df_lang = df[df["LANG_PRON"]==lang]
    lang_colex[lang] = df_lang.groupby(["COLEX"]).count()["SENSE_LEMMA"].to_dict()
```

```python
df_colex = pd.DataFrame.from_dict(lang_colex, orient="index")
```
