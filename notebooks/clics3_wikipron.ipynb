```python
import pandas as pd
```

```python
df = pd.read_csv("../data/colex/colex_clics3_glottolog.csv")
```

```python
langs = list(set(df["ISO639P3code"].tolist()))
```

```python
 len(langs)
```

```python
folder="../data/phon/wikipron/data/scrape/tsv"
```

```python
import os
langs_wiki = []
for file in os.listdir(folder):
    if file.endswith(".tsv"):
        lang = file.split("_")[0]
        langs_wiki.append(lang)
```

```python
len(langs_wiki)
```

```python
inter_langs = list(set(langs_wiki).intersection(set(langs)))
```

```python
df = df[df["ISO639P3code"].isin(inter_langs)]
```
