```python
pron_folder = "../data/phon/preprocessed"
```

```python
import glob, os
```

```python
lemma_dict = {}
for file in glob.glob(f"{pron_folder}/*.tsv"):
    print(file)
    lang = os.path.basename(file).replace(".tsv", "")
    df_pron = pd.read_csv(file, sep="\t")
    
    for lemma, pron in zip(df_pron["lemma"], df_pron["pron"]):
        lemma_dict[lemma+"_"+lang]=pron
        # df.loc[(df["SENSE_LEMMA"]==lemma) & (df["LANG_PRON"]==lang), "PRON"]=pron
```

```python
lang_lat = {}
lang_long = {}

for lang, lat, long in zip(df_langs["ISO639P3code"], df_langs["Latitude"], df_langs["Longitude"]):
    lang_lat[lang]=lat
    lang_long[lang] = long
    
```
